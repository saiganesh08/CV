```python

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("diamond.png")
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
 
sift=cv.SIFT_create()
detect = sift.detect(gray,None)
 
img = cv.drawKeypoints(gray,detect,img)
 
 
plt.imshow(img)
plt.title("SIFT")
plt.xticks([]),plt.yticks([])
plt.show()
```
