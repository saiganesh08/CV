```python
## Stastical Analysis 
import cv2 as cv
import numpy as np
 
img=cv.imread("beach.jpg")
img_array=np.asarray(img)
 
sum=np.sum(img_array)
avg=np.mean(img_array)
std=np.std(img_array)
min=np.min(img_array)
max=np.min(img_array)
 
print(" SUM:",sum)
print(" Average:",avg)
print(f" Standard Deviation: {std}")
print(f" Minimum: {min}")
print(f" Maximum: {max}")     
```

```python
## 4.Write a code to apply Different Filtering Operations on Images
import cv2 as cv
from matplotlib import pyplot as plt
img=cv.imread('sky.jpg')
plt.figure(figsize=(20,20))
 
g=cv.GaussianBlur(img,(5,5),0)
m=cv.medianBlur(img,7)
bi=cv.bilateralFilter(img,7,75,75)
box=cv.boxFilter(img,0,(7,7),(-1,-1))
 
l=[img,g,m,bi,box]
t=["Original","GaussianBlur","MedianBlur","BilateralFilter","BoxFilter"]
 
for i in range(5):
    plt.subplot(1,5,i+1)
    plt.imshow(l[i])
    plt.title(t[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
## 5.Write a code to apply morphological operations like dilation, erosion,opening and closing on the given image. 
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
plt.figure(figsize=(20, 20))
kernel=np.ones((5,5),np.uint8)
img=cv.imread('beach.jpg')
 
_,mask = cv.threshold(img, 150, 255, cv.THRESH_BINARY)
erosion=cv.erode(img,kernel,iterations=1)
dilation=cv.dilate(img,kernel,iterations=1)
opening=cv.morphologyEx(img,cv.MORPH_OPEN,kernel)
closing=cv.morphologyEx(img,cv.MORPH_CLOSE,kernel)
 
l=[img,erosion,dilation,opening,closing,mask]
t=["Original","Erosion","Dilation","Opening","Closing","Mask"]
 
for i in range(6):
    plt.subplot(1,6,i+1)
    plt.imshow(l[i])
    plt.title(t[i])
    plt.xticks([]),plt.yticks([])
plt.show() 
```

```python
## 7.Write a code to implement SIFT detector. 
 
import numpy as np
import cv2 as cv
 
img = cv.imread("diamond.png")

gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
 
sift=cv.SIFT_create()
detect = sift.detect(gray,None)
 
img = cv.drawKeypoints(gray,detect,img)
 
 
plt.imshow(img)
plt.title("SIFT")
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
## 8.Implement histogram calculation and equalization for the given image.
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
 
img = cv.imread('sky.jpg',0)
equalize = cv.equalizeHist(img)
res = np.hstack((img,equalize))
histogram_plot = cv.calcHist([img],[0],None,[250],[0,256])
 
plt.imshow(res)
plt.title("Image Histogram")
plt.xticks([]),plt.yticks([])
plt.show()

plt.plot (histogram_plot)
plt.title("Histogram")
plt.show()

```
