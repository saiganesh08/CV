```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

plt.figure(figsize=(10, 10))
img = cv2.imread("sky.jpg")
resize = cv2.resize(img,(220,200))
 
height,width = img.shape[:2]
center = (width/2, height/2)
rotate_matrix = cv2.getRotationMatrix2D(center=center, angle=25, scale=2)
rotated = cv2.warpAffine(src = img, M=rotate_matrix,dsize=(width,height))

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(height,width))
l=[img,resize,rotated,dst]
t=["Original","Resize","Rotated","Translation"]
 
for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(l[i])
    plt.title(t[i])
    plt.xticks([]),plt.yticks([])
plt.show()  
```
