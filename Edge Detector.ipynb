```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread("diamond.png",1)
```

```python
plt.figure(figsize=(30, 30))
 
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
 
laplacian=cv.Laplacian(img,cv.CV_64F)
canny=cv.Canny(img, 200, 250)
sobelx=cv.Sobel(img,cv.CV_64F,1,0,ksize=7)
sobely=cv.Sobel(img,cv.CV_64F,0,1,ksize=7)
prewittx = cv.filter2D(img, -1, kernelx)
prewitty = cv.filter2D(img, -1, kernely)
l=[img,laplacian,canny,sobelx,sobely,prewittx,prewitty]
t=["Original","Laplacian","Canny","Sobel X","Sobel Y","Perwitt X","Perwitt Y"]
 
for i in range(7):
    plt.subplot(3,3,i+1)
    plt.imshow(l[i])
    plt.title(t[i])
    plt.xticks([]),plt.yticks([])
plt.show()  
```
