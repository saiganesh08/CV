```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv.imread('diamond.png',0)
img = cv.resize(img,(300,300))
```

```python
equalize = cv.equalizeHist(img)
res = np.hstack((img,equalize))
histogram_plot = cv.calcHist([img],[0],None,[250],[0,256])
 
plt.imshow(res)
plt.title("Image Histogram")
plt.xticks([]),plt.yticks([])
plt.show()
plt.plot(histogram_plot)
plt.title("Histogram")
plt.show()
```
