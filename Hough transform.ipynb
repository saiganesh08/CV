```python
import cv2
from matplotlib import pyplot as plt
```

```python
original_img = cv2.imread('eclispe.jpg')
img          = original_img.copy()
gray_img     = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)



img_blur = cv2.medianBlur(gray_img, 5)
circles  = cv2.HoughCircles(
                img_blur,           # image
                cv2.HOUGH_GRADIENT, # method
                1,                  # dp
                80,                 # min distance : Minimum distance between the centers of the detected circles. If the parameter is
                                    # too small, multiple neighbor circles may be falsely detected in addition to a true one. If it is
                                    # too large, some circles may be missed.
                param1=50,          
                param2=30,
                minRadius=0,        # min radius
                maxRadius=0         # max radius
            )



for circle in circles[0,:]:
    cv2.circle(
        img,                                 # image
        (int(circle[0]),int(circle[1])),     # center
        int(circle[2]),                      # radius
        (220,255,0),                         # color
        3                                    # thickness
    )
    
    
    
    
plt.figure(figsize = (20, 10))

l = [original_img,img]
t = ["original image",'hough transformed image']

for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(cv2.cvtColor(l[i],cv2.COLOR_BGR2RGB))
    plt.title(t[i])

plt.show()
```
