```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize = (30,30))
img = cv2.imread('diamond.png', 0)
_,mask = cv2.threshold(img, 210, 255, cv2.THRESH_BINARY_INV)
kernal = np.ones((5,5), np.uint8)
dilation = cv2.dilate(mask, kernal, iterations = 1)
erode = cv2.erode(mask, kernal, iterations = 1)
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernal)
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernal)
titles = ['original image', 'masked image', 'dilation image', 'erosion image','opening image', 'closing image','gradient', 'tophat', 'blackhat']
images = [img, mask, dilation, erode,opening,closing]
for i in range(6):
          plt.subplot(1,6,i+1)
          plt.imshow(images[i], 'gray')
          plt.title(titles[i])
          plt.xticks([]), plt.yticks([])
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
